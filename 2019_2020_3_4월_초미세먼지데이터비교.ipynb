```python
import csv
years = ['2019','2020']
months = {'03':33,'04':32}

for year in years : 
    for month in months : 
        data = csv.reader(open('dustDetailInfo_'+year+'_'+month+'.csv', encoding = 'utf8'))
        next(data)

        dust = []
        for row in data : 
            for i in range(2, months[month]) : 
                if row[i] == '' :
                    row[i] = 0
                dust.append(int(row[i]))
        import matplotlib.pyplot as plt
        plt.title(year+' '+month)
        plt.boxplot(dust)
        plt.show()
```

```python
import csv
years = ['2019','2020']
months = {'03':33,'04':32}

for year in years : 
    for month in months : 
        data = csv.reader(open('dustDetailInfo_'+year+'_'+month+'.csv', encoding = 'utf8'))
        next(data)

        dust = []
        for row in data : 
            for i in range(2, months[month]) : 
                if row[i] == '' :
                    row[i] = 0
                dust.append(int(row[i]))
        import matplotlib.pyplot as plt
        plt.title(year+' '+month)
        plt.hist(dust)
        plt.show()
```

```python
import csv
years = ['2019','2020']
months = {'03':33,'04':32}
dust_all = []
for year in years : 
    for month in months : 
        data = csv.reader(open('dustDetailInfo_'+year+'_'+month+'.csv', encoding = 'utf8'))
        next(data)

        dust = []
        for row in data : 
            for i in range(2, months[month]) : 
                if row[i] == '' :
                    row[i] = 0
                dust.append(int(row[i]))
        dust_all.append(dust)
        
import matplotlib.pyplot as plt
plt.title('2019~2020, 3~4 Boxplot')
plt.boxplot(dust_all)
plt.show()
```
