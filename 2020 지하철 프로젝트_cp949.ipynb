```python
import csv
data = csv.reader(open('subway1903.csv', encoding = 'cp949'))
next(data)
next(data)

max_person1 = [0] * 24
max_station1 = [''] * 24

max_person2 = [0] * 24
max_station2 = [''] * 24
for row in data :
    for i in range(24):
        if int(max_person1[i]) < int(row[2+(i*2)]) :
            max_person1[i] = int(row[2+(i*2)])
            max_station1[i] = row[1] + '/' + str(i + 4) #+'('+ row[0]+')'
        if int(max_person2[i]) < int(row[3 + (i * 2)]):
            max_person2[i] = int(row[3 + (i * 2)])
            max_station2[i] = row[1] + '/' + str(i + 4)  # +'('+ row[0]+')'

for i in range(24):
    print(i+4,'시(승차): ', max_station1[i],'/',max_person1[i])
    print(i + 4,'시(하차): ', max_station2[i], '/', max_person2[i])

import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
plt.rc('font', family='Malgun Gothic')
cnt = np.arange(24)
plt.figure(dpi=150)
plt.title('2019년 3월 서울시 지하철 시간대별 최다 승차역')
plt.plot(cnt, max_person1,'r')
plt.xticks(cnt, max_station1,rotation='vertical')
plt.ylim(ymax=450000)
plt.savefig('2019년 3월최다승차역.png')
plt.show()
plt.figure(dpi=150)
plt.title('2019년 3월 서울시 지하철 시간대별 최다 하차역')
plt.plot(cnt, max_person2,'b')
plt.xticks(cnt, max_station2,rotation='vertical')
plt.ylim(ymax=450000)
plt.savefig('2019년 3월최다하차역.png')
plt.show()
```

```python
import csv
data = csv.reader(open('subway1903.csv', encoding = 'cp949'))
next(data)
next(data)
인원 = []
역이름 = input('무슨 역인가요?')
호선 = input('몇 호선인가요?')

for row in data :
    if(row[1] == 역이름 and row[0] == 호선):
        print(row[1])
        인원 = list(map(int, row[2:]))
    
승차 = 인원[::2]
하차 = 인원[1::2]

print(승차)
print(하차)

max_num1 = max(map(int,인원[::2]))
max_num2 = max(map(int,인원[1::2]))
max_num = max(max_num1, max_num2)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
labels = []
x = []
for i in range(4,28):
    labels.append(str(i)+'시')
    x.append(i)
    
plt.figure(dpi=200)
plt.xticks(x, labels,rotation='vertical')
plt.plot(x,승차,'r', label='승차 인원')
plt.plot(x,하차,'b', label='하차 인원')
plt.ylim(ymax=max_num + 4000)
plt.title(역이름+'('+호선+')')
plt.legend()
plt.savefig('test3.png')
plt.show()
```
